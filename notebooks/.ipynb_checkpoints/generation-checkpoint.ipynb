##### Generation 

```python
def get_one_hot_cond(cond_compact):

    cond = np.zeros((14), dtype='int')
    cond[cond_compact[0]-1] = 1
    cond[12+cond_compact[1]] = 1
    
    cond = cond.reshape(1,-1)
        
    return cond
```

```python
cond_compact = [1,0] # month, weekday or not
z = np.array([0,0]).reshape(1,-1) # projection on the latent space
```

```python
cond = get_one_hot_cond(cond_compact)
```

```python
x_generated = model.decoder.predict([z, cond])
```

```python
z.shape
```

```python
fig, axes = plt.subplots(4,3, figsize=(12,12))

ind = 0

for ax_row in axes:
    for ax in ax_row:
        
        cond_compact = [ind+1,1]
        z = np.array([0,0]).reshape(1,-1)
        cond = get_one_hot_cond(cond_compact)
       
        x_generated = model.decoder.predict([z, cond])
        x_generated_descale = (x_generated+scale_mean)*np.sqrt(scale_var)
        
        ax.plot(x_generated_descale[0,:47])
        ax.set_title('month: ' + str(ind+1))
        
        ind+=1

plt.tight_layout()
plt.show()
```

```python
# for one cond, latent space exploration

array_z = np.linspace(-2,2,12).reshape(-1,1)
array_z_1 = np.concatenate((np.zeros([12,1]), array_z), axis=1).reshape(-1,2)
array_z_2 = np.concatenate((array_z,np.zeros([12,1])), axis=1).reshape(-1,2)
array_z_3 = np.concatenate((array_z,array_z), axis=1).reshape(-1,2)
```

```python
cond_compact = cond_compact = [5,0] 
cond = get_one_hot_cond(cond_compact)
```

```python
for ind in range(array_z_1.shape[0]):

    z = array_z_1[ind,:].reshape(1,-1)

    x_generated = model.decoder.predict([z, cond])
    x_generated_descale = (x_generated+scale_mean)*np.sqrt(scale_var)

    plt.plot(x_generated_descale[0,:47], label='{:.2f}'.format(z[0,1]))

    ind+=1
    
plt.legend()
plt.show()

for ind in range(array_z_2.shape[0]):

    z = array_z_2[ind,:].reshape(1,-1)

    x_generated = model.decoder.predict([z, cond])
    x_generated_descale = (x_generated+scale_mean)*np.sqrt(scale_var)

    plt.plot(x_generated_descale[0,:47], label='{:.2f}'.format(z[0,0]))

    ind+=1
    
plt.legend()
plt.show()

for ind in range(array_z_3.shape[0]):

    z = array_z_3[ind,:].reshape(1,-1)

    x_generated = model.decoder.predict([z, cond])
    x_generated_descale = (x_generated+scale_mean)*np.sqrt(scale_var)

    plt.plot(x_generated_descale[0,:47], label='{:.2f}'.format(z[0,0]))

    ind+=1
    
plt.legend()
plt.show()
```

```python
# Latent space exploration around a given point 
ind = 1600
x = dataset['train']['x'][0][ind].reshape(1,-1)
cond = dataset['train']['x'][1][ind].reshape(1,-1)

z_proj = model.encoder.predict([x,cond])[0]
x_hat = model.decoder.predict([z_proj,cond])
```

```python
x_descale = (x+scale_mean)*np.sqrt(scale_var)
x_hat_descale = (x_hat+scale_mean)*np.sqrt(scale_var)
```

```python
r_max = 1
nb_points = 10
```

```python
r = np.random.uniform(high = r_max, size=nb_points)
theta = np.random.uniform(0, 2*np.pi, size=nb_points)

dz = np.concatenate((np.array(r*np.cos(theta)).reshape(nb_points,-1), 
                     np.array(r*np.sin(theta)).reshape(nb_points,-1)), axis=1)

z_array = z_proj + dz
```

```python
plt.plot(x_descale[0,:47], 'r', label = 'ori', lw=2)
plt.plot(x_hat_descale[0,:47], 'b',label = 'reconstru', lw=2)
plt.legend()
```
